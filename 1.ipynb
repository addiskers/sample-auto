````python
import json
from openai import OpenAI

api_key = "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"
client = OpenAI(api_key=api_key)


def generate_company_info(context: dict) -> dict:
    """
    Generate company information in structured JSON using OpenAI.

    Args:
        context (dict): Dictionary with keys "company_name" and "headline".

    Returns:
        dict: Company information filled with realistic details.
    """
    prompt = f"""Generate information about {context["company_name"]} in the "{context["headline"]}" domain. 
    Return the information in the following JSON format:
    {{
        "company_name": "{context["company_name"]}",
        "headquarters": "",
        "employee_count": "",
        "revenue": "",
        "top_product": "",
        "estd": "",
        "website": "",
        "geographic_presence": "",
        "ownership": "",
        "short_description_company": ""
    }}
    geographic_presence only choose between Region or Global
    Fill in realistic information for each field. The short_description_company should be around 150 words.
    revenue should be in the format "USD X billion" or "USD X million" and should be correct 2024 data.
    ownership should be either "Public" or "Private".
    estd is year of establishment should be correct data.
    Return ONLY valid JSON, no additional text."""

    try:
        response = client.responses.create(
            model="gpt-5",
            tools=[
                {
                    "type": "web_search_preview",
                    "search_context_size": "low",
                }
            ],
            input=[
                {
                    "role": "system",
                    "content": "You are a JSON generator. Always return valid JSON and nothing else.",
                },
                {"role": "user", "content": prompt},
            ],
        )
        content = response.output_text.strip()
        print("Raw response content:")
        print(content)
        # Clean code fences if present
        if content.startswith("```json"):
            content = content[7:]
        if content.endswith("```"):
            content = content[:-3]
        content = content.strip()

        try:
            return json.loads(content)
        except json.JSONDecodeError as json_error:
            print(f"JSON decode error: {json_error}")
            print(f"Response content: {content}")
            return {
                "company_name": context["company_name"],
                "headquarters": "",
                "employee_count": "",
                "revenue": "",
                "top_product": "",
                "estd": "",
                "website": "",
                "geographic_presence": "",
                "ownership": "",
                "short_description_company": "",
            }

    except Exception as api_error:
        print(f"OpenAI API error: {api_error}")
        return {
            "company_name": context["company_name"],
            "headquarters": "",
            "employee_count": "",
            "revenue": "",
            "top_product": "",
            "estd": "",
            "website": "",
            "geographic_presence": "",
            "ownership": "",
            "short_description_company": "",
        }


context = {"company_name": "Beijing didi", "headline": "Robo taxi market "}

result = generate_company_info(context)
result
````
